## Classificação dos Times no Brasileirão 2022

Começamos importando as bibliotecas que vamos utilizar.

```python
import numpy as np
import pandas as pd
import re
```

Agora importamos o nosso arquivo csv com os dados do Campeonato Brasileiro 2022 como um pandas.DataFrame, e substituímos/retiramos os caracteres especiais (vogais acentuadas) e informações desnecessárias afim de reaprovaitar esses dados em em outra aplicação.

```python
df = pd.read_csv("resultados.csv")
for i in range(20):
    df[" "][i] = re.sub("é", "e", df[" "][i])
    df[" "][i] = re.sub("í", "i", df[" "][i])
    df[" "][i] = re.sub("á", "a", df[" "][i])
    df[" "][i] = re.sub("RB ", "", df[" "][i])
    df[" "][i] = re.sub("ã", "a", df[" "][i])
```

Fizemos uma matriz de 0's que servirá de base, onde os 0's serão substituídos dependendo do resultado de cada jogo. 

```python
matriz_resultados = [[0 for i in range(20)] for j in range(20)]
matriz_resultados
```

A lista "jogos" carregará quantos partidas cada time jogou, de início definimos 40 partidas para todos os times (estamos considerando um time jogando contra ele mesmo) 

```python
jogos = [40 for i in range(20)]
```

No nosso csv, temos o símbolo "—" na diagonal principal da matriz, que é quando temos a relação de um time com ele mesmo, e possuímos o formato "R–40" que representa em qual rodada a partida será jogada. Então subtraímos 1 na lista "jogos", do time-coluna e do time-linha quando não houve jogo entre eles, no caso do "—" é retirado duas vezes, pois a linha e a coluna é o mesmo time, corrigindo as 40 partidas que definimos antes. Para as células que não possuem "—" e o formato "R-40", nós temos o formato "5–2", onde o primeiro número representa os gols do time-linha (casa) e o segundo do time-coluna (fora), quando o a quantidade de gols da casa for maior que a de fora, então somamos 1 na célula (casa, fora), do contrário somamos 1 na célula (fora, casa), e nos casos em que nenhum for maior (empate) somamos 0.5 em ambas as células.

```python
i = 0
for coluna in df:
    if coluna == " ":
        continue
    for time in range(20):
        resultado = df.at[time, coluna]
        if resultado[0] in "—R":
            jogos[time] -= 1
            jogos[i] -= 1
            continue
        casa, visitante = resultado.split("–")
        if casa > visitante:
            matriz_resultados[time][i] += 1
        elif casa < visitante:
            matriz_resultados[i][time] += 1
        else:
            matriz_resultados[time][i] += 0.5
            matriz_resultados[i][time] += 0.5
    i += 1
matriz_resultados
```

Após a retirada dos jogos que não aconteceram, chegamos a seguinte lista.

```python
jogos
```

Como vimos na nossa lista "jogos", tem times que tiveram mais jogos que outros, e isso leva eles a conseguir uma maior soma de pontos, por isso, dividimos cada elemento da linha pela quantidade de partidas que o time-linha teve.

```python
matriz_preferencia = np.array(matriz_resultados)
for i in range(20):
    matriz_preferencia[i] /= jogos[i]
matriz_preferencia 
```

Utilizamos np.linalg.eig para calcular os autovalores e a matriz de autovetores correspondentes, como ele retorna os autovalores em ordem decrescente, e queremos o autovetor do maior autovalor, então selecionamos apenas a primeira coluna da matriz de autovetores. 

```python
val, vet = np.linalg.eig(matriz_preferencia)
pontuacoes = vet[:,0]
```

Criamos uma lista com os nomes do times.

```python
lista_times = list(df[" "])
```

E por fim, reordenamos os elementos da lista de times seguindo a regra do maior valor para o menor do nosso autovetor.

```python
ranking = [x for _, x in sorted(zip(pontuacoes, lista_times))]
ranking = ranking[::-1]
ranking
```

## Ranking Histórico

Podemos aumentar nossa base de dados para tornar o ranking mais abrangente. Note que aqui usamos o método direto melhorado (com a função h(x)).

Agradecimentos a Adão Duque pela compilação de dados do Campeonato Brasileiro (https://github.com/adaoduque/Brasileirao_Dataset).

```python
df = pd.read_csv("2003_2021.csv")
df = df.drop(columns=["rodada", "ID", "hora", "data", "dia", "formacao_mandante", "formacao_visitante", "tecnico_mandante", "tecnico_visitante", "vencedor", "arena", "mandante_estado", "visitante_estado", "estado_vencedor"])
df2022 = pd.read_csv("resultados.csv")
for i in range(20):
    df2022[" "][i] = re.sub("é", "e", df2022[" "][i])
    df2022[" "][i] = re.sub("í", "i", df2022[" "][i])
    df2022[" "][i] = re.sub("á", "a", df2022[" "][i])
    df2022[" "][i] = re.sub("RB ", "", df2022[" "][i])
    df2022[" "][i] = re.sub("ã", "a", df2022[" "][i])
    df2022[" "][i] = re.sub("fogo", "fogo-RJ", df2022[" "][i])

lista_times = list(set(df["mandante"]))
lista_times22 = list(df2022[" "])

N = len(lista_times)
matriz_preferencia = [[0 for i in range(N)] for j in range(N)]
jogos = [0 for i in range(N)]

traduz = [lista_times.index(lista_times22[i]) for i in range(20)]

def sgn(x):
    if x > 0:
        return 1
    if x == 0:
        return 0
    return -1

def h(x):
    ret = 0.5 + 0.5*sgn(x-0.5)*(abs(2*x-1)**0.5)
    return ret

for i in range(7645):
    a = lista_times.index(df.at[i, "mandante"])
    b = lista_times.index(df.at[i, "visitante"])
    jogos[a] += 1
    jogos[b] += 1
    casa, fora = df.at[i, "mandante_placar"], df.at[i, "visitante_placar"]
    pont_a = (casa+1)/(casa+fora+2)
    pont_b = (fora+1)/(casa+fora+2)
    matriz_preferencia[a][b] += h(pont_a)
    matriz_preferencia[b][a] += h(pont_b)

i = 0
for coluna in df2022:
    if coluna == " ":
        continue
    for time in range(20):
        resultado = df2022.at[time, coluna]
        if resultado[0] in "—R":
            continue
        a = traduz[time]
        b = traduz[i]
        
        jogos[a] += 1
        jogos[b] += 1
        
        casa, fora = resultado.split("–")
        casa, fora = int(casa), int(fora)

        pont_a = (casa+1)/(casa+fora+2)
        pont_b = (fora+1)/(casa+fora+2)

        matriz_preferencia[a][b] += h(pont_a)
        matriz_preferencia[b][a] += h(pont_b)
    i += 1

matriz_preferencia = np.array(matriz_preferencia)

for i in range(N):
    matriz_preferencia[i] /= jogos[i]

val, vet = np.linalg.eig(matriz_preferencia)
pontuacoes = vet[:,0]

ranking = [x for _, x in sorted(zip(pontuacoes, lista_times))]
ranking = ranking[::-1]

ranking
```

## Prevendo a Copa do Mundo

Nessa seção, utilizamos uma base dados que reúne todas as partidas entre as seleções da Copa do Mundo 2022 desde 2018 (após a última copa). Aqui, utilizamos o método probabilístico.

```python
df_copa = pd.read_csv("Copa_Historico.CSV", delimiter="|")
df_copa
```

```python
matriz_aux = [[0 for i in range(32)] for j in range(32)]

lista_selecoes = list(set(df_copa["Casa"]))
lista_selecoes.append("Uruguai")

for i in range(253):
    a = lista_selecoes.index(df_copa.at[i, "Casa"])
    b = lista_selecoes.index(df_copa.at[i, "Fora"])
    gol_casa = df_copa.at[i, "Casa_placar"]
    gol_fora = df_copa.at[i, "Fora_placar"]
    G_a = (gol_casa+1)/(gol_casa+gol_fora+2)
    G_b = 1 - G_a
    matriz_aux[a][b] += h(G_a)
    matriz_aux[b][a] += h(G_b)

matriz_copa = [[0 for i in range(32)] for j in range(32)]
for i in range(32):
    for j in range(32):
        if i == j:
            continue
        matriz_copa[i][i] += (matriz_aux[j][i]**2)

for i in range(32):
    for j in range(32):
        if i == j:
            continue
        matriz_copa[i][j] = -matriz_aux[i][j]*matriz_aux[j][i]

matriz_copa = np.array(matriz_copa)

diag = max(np.amax(matriz_copa), -np.amin(matriz_copa)) + 1

matriz_copa = matriz_copa + np.eye(32)*diag

val, vet = np.linalg.eig(matriz_copa)

indice_max = -1
autovalor_max = -1

for i in range(32):
    if np.linalg.norm(val[i]) < autovalor_max or autovalor_max == -1:
        autovalor_max = np.linalg.norm(val[i])
        indice_max = i

pontuacoes = vet[:,indice_max]

ranking = [(x, _) for _, x in sorted(zip(pontuacoes, lista_selecoes))]
ranking = ranking[::-1]

ranking
```

```python
gA = ['Catar', 'Equador', 'Senegal', 'Holanda']
gB = ['Inglaterra', 'Ira', 'Estados Unidos', 'Gales']
gC = ['Argentina', 'Arabia Saudita', 'Mexico', 'Polonia']
gD = ['Franca', 'Australia', 'Dinamarca', 'Tunisia']
gE = ['Espanha', 'Costa Rica', 'Alemanha', 'Japao']
gF = ['Belgica', 'Canada', 'Marrocos', 'Croacia']
gG = ['Brasil', 'Servia', 'Suica', 'Camaroes']
gH = ['Portugal', 'Gana', 'Uruguai', 'Coreia do Sul']
grupos = [[[ranking[i][0] for i in range(32)].index(k[j]) for j in range(4)] for k in [gA, gB, gC, gD, gE, gF, gG, gH]]
grupos
```

```python
def prob(i, j):
    if i == j:
        return 1
    return ranking[i][1]/(ranking[i][1]+ranking[j][1])
```

```python
prob_grupos = [[0, 0] for i in range(32)]
for grupo in grupos:
    partidas = [[0,1],[2,3],[0,2],[1,3],[0,3],[1,2]]
    for mask in range(64):
        pont = [0, 0, 0, 0]
        P = 1
        for bit in range(6):
            if mask & (1 << bit):
                P *= prob(grupo[partidas[bit][0]],grupo[partidas[bit][1]])
                pont[partidas[bit][0]] += 1
            else:
                P *= prob(grupo[partidas[bit][1]],grupo[partidas[bit][0]])
                pont[partidas[bit][1]] += 1
        x = sorted([[-pont[i], grupo[i]] for i in range(4)])
        prob_grupos[x[0][1]][0] += P
        prob_grupos[x[1][1]][1] += P
```

```python
prob_oitavas = [[0,0] for i in range(32)]

for grupo_i in range(8):
    grupo = grupos[grupo_i]

    for pos in [0, 1]:
        for i in grupo:
            grupo_j = grupo_i + (-1)**(grupo_i % 2)
            for j in grupos[grupo_j]:
                prob_oitavas[i][(grupo_i-pos)%2] += prob_grupos[i][pos]*prob_grupos[j][1-pos]*prob(i,j)

prob_quartas = [[0,0] for _ in range(32)]

for grupo_i in range(8):
    grupo = grupos[grupo_i]

    grupo_j_min = 4*(grupo_i//4)+{0:2, 1:2, 2:0, 3:0}[grupo_i%4]

    for i in grupo:
        for grupo_j in range(grupo_j_min, grupo_j_min+2):
            for lado in [0, 1]:
                for j in grupos[grupo_j]:
                    prob_quartas[i][lado] += prob_oitavas[i][lado]*prob_oitavas[j][lado]*prob(i,j)

prob_semi = [[0,0] for _ in range(32)]

for grupo_i in range(8):
    grupo = grupos[grupo_i]

    if grupo_i < 4:
        grupo_j_min = 4
    else:
        grupo_j_min = 0

    for i in grupo:
        for grupo_j in range(grupo_j_min, grupo_j_min+4):
            for lado in [0, 1]:
                for j in grupos[grupo_j]:
                    prob_semi[i][lado] += prob_quartas[i][lado]*prob_quartas[j][lado]*prob(i,j)

prob_final = [0 for _ in range(32)]

for i in range(32):
    for j in range(32):
        if i == j:
            continue
        prob_final[i] += prob_semi[i][0]*prob_semi[j][1]*prob(i,j)
        prob_final[j] += prob_semi[i][0]*prob_semi[j][1]*prob(j,i)

sum([prob_final[i] for i in range(32)])
```
